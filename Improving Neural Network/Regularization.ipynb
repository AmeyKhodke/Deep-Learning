```python
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import make_moons

import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
X,y = make_moons(100,noise=0.25,random_state=2)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
# Artificial Neural Network
model = Sequential()
model.add(Dense(128,activation='relu',input_dim=2))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
from keras.optimizers import Adam
adam = Adam(learning_rate=0.01)
model.compile(optimizer=adam,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X,y,epochs=2000,validation_split=0.2,verbose=1)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
# Plot decision boundary
plot_decision_regions(X,y,clf=model,legend=2)
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()
```

```python
# Model with L2 Regularization
model2 = Sequential()
model2.add(Dense(128,activation='relu',input_dim=2,kernel_regularizer=tensorflow.keras.regularizers.l2(0.01)))
model2.add(Dense(128,activation='relu',kernel_regularizer=tensorflow.keras.regularizers.l2(0.01)))
model2.add(Dense(1,activation='sigmoid'))
```

```python
adam = Adam(learning_rate=0.01)
model2.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
```

```python
history2 = model2.fit(X,y,epochs=2000,validation_split=0.2,verbose=1)
```

```python
plt.plot(history2.history['accuracy'])
plt.plot(history2.history['val_accuracy'])
```

```python
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
```

```python
plot_decision_regions(X,y,clf=model2,legend=2)
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()
```
