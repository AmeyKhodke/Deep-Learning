```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame([[8,8,1],[7,9,1],[6,10,0],[5,5,0]],columns=['cgpa','profile_score','placement'])
```

```python
def initialize_parameters(layer_dims):
  np.random.seed(3)
  L = len(layer_dims) # no. of layers in neural network
  parameters = {}
  for l in range(1,L):
    parameters['W'+str(l)] = np.ones((layer_dims[l-1],layer_dims[l]))*0.01
    parameters['b'+str(l)] = np.zeros((layer_dims[l],1))
  return parameters
```

```python
initialize_parameters([2,2,1])
```

```python
def sigmoid(Z):
  A = 1/(1+np.exp(-Z))
  return A
```

```python
def linear_forward(A_prev,w,b):
  Z = np.dot(w.T , A_prev)+b
  A = sigmoid(Z)
  return A
```

```python
# Forward Propogation
def forward_propogation(X,parameters):
  A = X
  L = len(parameters) // 2
  for l in range(1,L+1):
     A_prev = A
     wl = parameters['W'+str(l)]
     bl = parameters['b'+str(l)]
     print("A"+str(l-1),": ",A_prev)
     print("W"+str(l),": ",wl)
     print("b"+str(l),": ",bl)
     print("--"*20)
     A = linear_forward(A_prev,wl,bl)
  return A,A_prev
```

```python
X = data[['cgpa','profile_score']].values[0].reshape(2,1)
y = data[['placement']].values[0][0]
parameters = initialize_parameters([2,2,1])
y_hat,A1 = forward_propogation(X,parameters)
```

```python
X = data[['cgpa','profile_score']].values[0].reshape(2,1)
y = data[['placement']].values[0][0]
param = initialize_parameters([2,2,1])
y_hat,A1 = forward_propogation(X,param)
y_hat = y_hat[0][0]
update_parameters(param,y,y_hat,A1,X)
loss = -y * np.log(y_hat) - (1-y) * np.log(1-y_hat)
print("Loss for student: ",loss)
parameters
```
